```python
# 🛒 Superstore Sales Analysis

## 📌 Project Goal
#Analyze sales and profitability data to:
#- Identify top-performing categories
#- Detect loss-making segments
#- Understand regional performance
#- Discover trends over time

# ===============================
# 1. Import Libraries
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional styling
sns.set(style="whitegrid")
```

```python

# ===============================
# 2. Load Dataset
# ===============================
df = pd.read_csv("Sample - Superstore.csv" , encoding="latin1")
df.head()
```

```python
# ===============================
# 3.data cleaning
# ===============================

df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month

print(df.isnull().sum())
print("Duplicates:", df.duplicated().sum())
```

```python
# ===============================
# 4.kpis
# ===============================
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_orders = df["Order ID"].nunique()
profit_margin = (total_profit / total_sales) * 100

print("Total Sales:", round(total_sales, 2))
print("Total Profit:", round(total_profit, 2))
print("Total Orders:", total_orders)
print("Profit Margin (%):", round(profit_margin, 2))
```

```python

# ===============================
# 5. sales by category
# ===============================
df.groupby("Category")["Sales"].sum().plot(kind="bar", title="Sales by Category")
plt.show()
```

```python
# ===============================
# 6. profit by category
# ===============================
df.groupby("Category")["Profit"].sum().plot(kind="bar", title="Profit by Category", color="orange")
plt.show()
```

```python
# ===============================
# 7. sales by region
# ===============================
df.groupby("Region")["Sales"].sum().plot(kind="bar", title="Sales by Region")
plt.show()
```

```python

# =====================================
# 8 disc vs profit
# =====================================
sns.scatterplot(data=df, x="Discount", y="Profit")
plt.title("Discount vs Profit")
plt.show()
```

```python
# =====================================
# 9. monthly sales trend
# =====================================
df.set_index("Order Date", inplace=True)

df["Sales"].resample("ME").sum().plot(figsize=(10,5))
plt.title("Sales Over Time")
plt.show()
```

```python
# ===============================
# customer Segmentation
# ===============================
customer_sales = df.groupby("Customer Name")["Sales"].sum()
segments = pd.qcut(customer_sales, q=3, labels=["Low", "Medium", "High"])

segments.value_counts().plot(kind="bar", title="Customer Segmentation")
plt.show()
```

```python
# =====================================
# 11 Dashboard View
# =====================================

fig, axes = plt.subplots(2, 2, figsize=(14,10))

df.groupby("Category")["Sales"].sum().plot(kind="bar", ax=axes[0,0], title="Sales by Category")
df.groupby("Region")["Profit"].sum().plot(kind="bar", ax=axes[0,1], title="Profit by Region")
df.groupby("Month")["Sales"].sum().plot(ax=axes[1,0], title="Monthly Sales")
df.groupby("Segment")["Profit"].sum().plot(kind="bar", ax=axes[1,1], title="Profit by Segment")

plt.tight_layout()
plt.show()
```

```python
# 📖 Business Summary

#- Technology drives most revenue and profit  
#- High discounts reduce profitability  
#- Some products generate consistent losses  
#- A small group of customers drives most sales  

## 🚀 Recommendations
#- Focus on high-performing categories  
#- Reduce discounts  
#- Improve or remove low-performing products  
#- Retain high-value customers  
```
