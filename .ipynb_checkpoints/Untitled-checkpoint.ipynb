```python
# ===============================
# 1. Import Libraries
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Optional styling
sns.set(style="whitegrid")
```

```python

# ===============================
# 2. Load Dataset
# ===============================

df = pd.read_csv("Sample - Superstore.csv", encoding="latin1")
df.head()
```

```python
# ===============================
# 3.Convert dates
# ===============================

df["Order Date"] = pd.to_datetime(df["Order Date"])
df["Ship Date"] = pd.to_datetime(df["Ship Date"])

# Check missing values
print("\nMissing Values:\n", df.isnull().sum())

# Check duplicates
print("\nDuplicate Rows:", df.duplicated().sum())

# Drop duplicates
df = df.drop_duplicates()
```

```python

# ===============================
# 4. Dataset overview
# ===============================

print("Shape:", df.shape)
print("Columns:", df.columns)

print("Date Range:",
      df["Order Date"].min(),
      "to",
      df["Order Date"].max())

print("Unique Customers:", df["Customer ID"].nunique())
print("Unique Products:", df["Product ID"].nunique())
```

```python


# ===============================
# 5. Explore the Dataset
# ===============================

# First rows
df.head()

# Dataset structure
df.info()

# Basic statistics
df.describe()
```

```python
# ===============================
# 6. quality check
# ===============================

print("Negative profit orders:", (df["Profit"] < 0).sum())
print("\nWorst Orders (Loss-Making)")


print("Zero sales:", (df["Sales"] == 0).sum())

print("High discount orders (>50%):", (df["Discount"] > 0.5).sum())
```

```python
# ===============================
# 7. Basic Metrics
# ===============================

# Total sales
df["Sales"].sum()

# Average profit
df["Profit"].mean()

# Maximum sale
df["Sales"].max()

# Total Profit
total_profit = df["Profit"].sum()

# Profit Margin
profit_margin = total_profit / df["Sales"].sum()

# Average Discount
avg_discount = df["Discount"].mean()

print(total_profit, profit_margin, avg_discount)

```

```python

# =====================================
# 8 Sales Analysis
# =====================================
# ===============================
# 9 Data Visualization
# ===============================
plt.figure(figsize=(8,5))
sales_by_region = df.groupby("Region")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Region", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize=(8,5))
df.groupby("Category")["Sales"].sum().plot(kind="bar")
plt.title("Sales by Category", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# ===============================
# 10 Profit by Category 
# ===============================
df.groupby("Category")["Profit"].sum().plot(kind="bar")

plt.title("Profit by Category", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# Discount vs Profit
plt.scatter(df["Discount"], df["Profit"])
plt.xlabel("Discount")
plt.ylabel("Profit")
plt.title("Discount vs Profit", fontsize=14)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
# =====================================
# 11 Profit Analysis
# =====================================

profit_by_category = df.groupby("Category")["Profit"].sum()

profit_by_subcategory = df.groupby("Sub-Category")["Profit"].sum().sort_values(ascending=False)

print("\nProfit by Category")
print(profit_by_category)

print("\nTop Sub-Categories by Profit")
print(profit_by_subcategory.head(10))
```

```python
# =====================================
# 12 Sorting Example
# =====================================

top_sales_orders = df.sort_values("Sales", ascending=False).head(10)

print("\nTop Sales Orders")
print(top_sales_orders)
```

```python

# =====================================
# 13 Identify Loss-Making Orders
# =====================================

loss_orders = df[df["Profit"] < 0]

print("\nWorst Orders (Loss-Making)")
print(loss_orders.sort_values("Profit").head(10))
```

```python
# ===============================
# 14. Sorting Data
# ===============================

# Highest profit orders
df.sort_values("Profit", ascending=False)

# Lowest profit orders
df.sort_values("Profit")
```

```python
# ===============================
# 15. Filtering Data
# ===============================

# Orders with sales greater than 1000
df[df["Sales"] > 1000]

# Technology category only
df[df["Category"] == "Technology"]

# High profit orders
df[df["Profit"] > 500]
```

```python
# ===============================
# 16. Create New Columns
# ===============================

# Profit ratio
df["Profit Ratio"] = df["Profit"] / df["Sales"]
```

```python
# =====================================
# 17 Multiple Charts Dashboard
# =====================================

fig, axes = plt.subplots(2,2, figsize=(15,12))

# Sales by Region
df.groupby('Region')['Sales'].sum().plot(kind='bar', ax=axes[0,0], color='skyblue')
axes[0,0].set_title('Total Sales by Region')
axes[0,0].tick_params(axis='x', rotation=45)

# Profit by Category
df.groupby('Category')['Profit'].sum().plot(kind='bar', ax=axes[0,1], color='lightgreen')
axes[0,1].set_title('Total Profit by Category')
axes[0,1].tick_params(axis='x', rotation=45)

# Ship Mode Distribution
df.groupby('Ship Mode')['Order ID'].count().plot(kind='pie', ax=axes[1,0], autopct='%1.1f%%')
axes[1,0].set_title('Order Distribution by Ship Mode')
axes[1,0].set_ylabel('')

# Segment Sales
df.groupby('Segment')['Sales'].sum().plot(kind='bar', ax=axes[1,1], color='orange')
axes[1,1].set_title('Sales by Customer Segment')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()

plt.show()
```

```python
# =====================================
# 18 Time-Based Analysis
# =====================================

df["Order Date"] = pd.to_datetime(df["Order Date"])

df["Year"] = df["Order Date"].dt.year

df["Month"] = df["Order Date"].dt.month

df["Quarter"] = df["Order Date"].dt.quarter


# Monthly Sales Trend
monthly_sales = df.groupby(df["Order Date"].dt.to_period("M"))["Sales"].sum()

plt.figure(figsize=(12,6))

monthly_sales.plot(marker="o")

plt.title("Monthly Sales Trend")

plt.xlabel("Month")

plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.grid(True)

plt.show()
```

```python
sns.heatmap(df[["Sales", "Profit", "Discount", "Quantity"]].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
